```python
import pandas as pd
import time
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Load the dataset
df = pd.read_csv('bda.csv')

# Define the target variable and class labels
target_variable = "disability_type"
class_labels = df[target_variable].unique()

# Calculate the average of age
df['age_average'] = (df['age_lower'] + df['age_upper']) / 2

# Training and testing data
X = df[['age_average']]  # Use a DataFrame for X
y = df[target_variable]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predict probabilities for the test set
prob_predict = dt_model.predict_proba(X_test)

# Display the probabilities for the first sample in the test set
print(prob_predict[0])

# If you want to see the class labels along with their probabilities for the first sample:
class_probabilities = list(zip(dt_model.classes_, prob_predict[0]))
print(class_probabilities)

# Measure the training time
start_time = time.time()
dt_model.fit(X, y)
end_time = time.time()

training_time = end_time - start_time
print(f"Training time: {training_time} seconds")


# Make predictions on the test set
y_pred = dt_model.predict(X_test)

# Calculate and display evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Display the metrics
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)

# Save the metrics to a file
with open('evaluation_metrics.txt', 'w') as file:
    file.write(f"Accuracy: {accuracy}\n")
    file.write("Confusion Matrix:\n")
    file.write(str(conf_matrix) + '\n')
    file.write("Classification Report:\n")
    file.write(classification_rep)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Load the dataset
df = pd.read_csv('bda.csv')

# Define the target variable and class labels
target_variable = "disability_type"

# Calculate the average of age
df['age_average'] = (df['age_lower'] + df['age_upper']) / 2

# Training and testing data with stratified sampling
X = df[['age_average']]
y = df[target_variable]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Create a Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = dt_model.predict(X_test)

# Calculate and display evaluation metrics with zero_division parameter
classification_rep = classification_report(y_test, y_pred, target_names=df[target_variable].unique(), zero_division=1)

# Display the metrics
print("Classification Report:")
print(classification_rep)

# Save the metrics to a file
with open('classification_report.txt', 'w') as file:
    file.write("Classification Report:\n")
    file.write(classification_rep)
```
